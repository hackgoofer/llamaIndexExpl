# Get Documents

```python
from llama_index.core import Document
import glob
import os

documents = []

def process_file(filename):
    with open(filename, 'r') as f:
        stripped_file_path = "".join(filename[60:].split(".")[:-1])
        section_title = stripped_file_path.split("/")[0]
        print(section_title)
        text = f.read()
        documents.append(
            Document(
                text=text,
                metadata={
                    "filename": stripped_file_path,
                    "section_title": section_title,
                    "section_list":  ""
                }
            )
        )

def process_directory(path):
    for filename in glob.glob(os.path.join(path, '*')):
        if filename.startswith(("_", ".")):
            continue
        if os.path.isdir(filename):
            process_directory(filename)
        else:
            process_file(filename)

process_directory("/Users/sasha/github/LlamaIndex/llama_index/docs/_build/text")
```
